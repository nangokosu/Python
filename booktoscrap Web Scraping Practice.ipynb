```python
import scrapy
```

```python
from collections import defaultdict
book_details=defaultdict(list)
class testcrawler2(scrapy.Spider):
    name="testcrawler2"
    start_urls=["http://books.toscrape.com/catalogue/page-1.html"]    
    def parse(self,response):
        for book in response.css("article.product_pod"):
            book_details["book name"].append(book.css("h3>a::attr(title)").extract()[0])
            book_details["book price"].append(book.css("div.product_price>p.price_color::text").extract()[0])
            book_details["book ratings"].append(book.css("p:nth-of-type(1)::attr(class)")[0].extract())
        for a in response.css('li.next a'):
            yield response.follow(a, self.parse)
        
```

```python
from scrapy.crawler import CrawlerProcess
process=CrawlerProcess({'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)'})
process.crawl(testcrawler2)
process.start()
```

```python
print(book_details)
```

```python
import pandas as pd
```

```python
book_details["book ratings"]=[string.split(sep=" ")[-1] for string in book_details["book ratings"]]
book_details["book price"]=[string[1:] for string in book_details["book price"]]
book_list=pd.DataFrame(book_details)
book_list["book price"]=pd.to_numeric(book_list["book price"])

```

```python
book_list["book ratings"]=pd.Categorical(book_list["book ratings"],ordered=True,categories=["One","Two","Three","Four","Five"])
print(book_list.info())
```

```python
grouped_mean=book_list.groupby("book ratings",as_index=False).agg({"book price":['mean','median']})
grouped_mean['mean_price']=grouped_mean["book price"]["mean"]
grouped_mean['median_price']=grouped_mean["book price"]["median"]
print(grouped_mean)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
pricevsrating=sns.catplot(x="book ratings",y="book price",data=book_list,kind="violin")
pricevsrating.set(xlabel="Star Ratings",ylabel="Price in euros")
plt.show()
```

```python
grouped_mm=pd.melt(grouped_mean,id_vars="book ratings",value_vars=["median_price","mean_price"],var_name="metric",value_name="Value")
print(grouped_mm)
mm_drawing=sns.catplot(x="book ratings",y="Value",data=grouped_mm,hue="metric",kind="point")
plt.show()
```
